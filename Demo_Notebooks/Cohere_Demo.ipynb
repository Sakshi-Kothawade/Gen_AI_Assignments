```python
from langchain_cohere import ChatCohere
from langchain_core.messages import HumanMessage

llm = ChatCohere()
messages = [
    HumanMessage(
        content="What should I include in my diet to stay energetic and make a productive day?"
    )
]
result = llm.invoke(messages)
print(result.content)
```

```python
prompt = """
What is sentiment of this review. Give one word answer p or n.
The food was tasty and the service was good. I had good experience of the hotel.
"""
result=llm.invoke(prompt)
print(result)
```
