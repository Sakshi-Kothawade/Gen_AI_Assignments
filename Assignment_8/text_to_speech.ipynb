```python
import torch
import soundfile as sf
from IPython.display import Audio
from espnet2.bin.tts_inference import Text2Speech
```

```python
text = "Good morning everyone. This project demonstrates text to speech synthesis using deep learning models."
```

# **MODEL 1 — TACOTRON2 + WAVENET**

```python
tacotron = Text2Speech.from_pretrained(
    model_tag="kan-bayashi/ljspeech_tacotron2",
    device="cuda" if torch.cuda.is_available() else "cpu"
)
```

```python
import nltk

nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt')
nltk.download('cmudict')
```

```python
with torch.no_grad():
    wav = tacotron(text)["wav"]

sf.write("tacotron.wav", wav.cpu().numpy(), tacotron.fs)
Audio("tacotron.wav")
```

# **MODEL 2 — FASTSPEECH2 (Parallel TTS)**

```python
fastspeech = Text2Speech.from_pretrained(
    model_tag="kan-bayashi/ljspeech_fastspeech2",
    device="cuda" if torch.cuda.is_available() else "cpu"
)
```

```python
with torch.no_grad():
    wav = fastspeech(text)["wav"]

sf.write("fastspeech.wav", wav.cpu().numpy(), fastspeech.fs)
Audio("fastspeech.wav")
```

# **MODEL 3 — VITS**

```python
vits = Text2Speech.from_pretrained(
    model_tag="kan-bayashi/ljspeech_vits",
    device="cuda" if torch.cuda.is_available() else "cpu"
)
```

```python
with torch.no_grad():
    wav = vits(text)["wav"]

sf.write("vits.wav", wav.cpu().numpy(), vits.fs)
Audio("vits.wav")
```

# **MODEL 4 — GRIFFIN-LIM (CLASSICAL VOCODER)**

```python
griffin = Text2Speech.from_pretrained(
    model_tag="kan-bayashi/ljspeech_tacotron2",
    vocoder_tag=None,
    device="cuda" if torch.cuda.is_available() else "cpu"
)
```

```python
with torch.no_grad():
    wav = griffin(text)["wav"]

sf.write("griffin_lim.wav", wav.cpu().numpy(), griffin.fs)
Audio("griffin_lim.wav")
```

## **LLM → Speech Pipeline**

```python
from transformers import pipeline
import re

llm = pipeline(
    task="text-generation",
    model="distilgpt2"
)

prompt = (
    "Write a short spoken introduction for a college AI project presentation. "
    "Do not mention websites, links, videos, apps, or platforms. "
    "Use simple spoken English."
)
```

```python
result = llm(
    prompt,
    max_new_tokens=50,
    do_sample=True,
    temperature=0.7,
    top_p=0.9,
    repetition_penalty=1.3,
    return_full_text=False
)

raw_text = result[0]["generated_text"]
print("Raw Output:\n", raw_text)
```

```python
def clean_text(text):
    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)
    # Remove repeated dots or symbols
    text = re.sub(r"[^\w\s,.]", "", text)
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text)
    return text.strip()

generated_text = clean_text(raw_text)

print("\nCleaned Text:\n", generated_text)
```

```python
with torch.no_grad():
    wav = vits(generated_text)["wav"]

sf.write("final_genai.wav", wav.cpu().numpy(), vits.fs)
Audio("final_genai.wav")
```

| VITS | Very High | Fast | End-to-End |
| Griffin-Lim | Low | Fast | Traditional DSP |
