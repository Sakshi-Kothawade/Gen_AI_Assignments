# **Promt Engineering Approaches**

## 1. **Zero Shot Prompting:**
The model answers the task without examples. It relies only on pretrained knowledge and may produce less accurate responses for complex tasks.

## **2. Few Shot Prompting:**
Examples are provided before the task. This helps the model understand the pattern and improves accuracy compared to zero-shot prompting.

## **3. Chain of Thought (CoT):**
The model is encouraged to reason step by step. This approach improves performance on logical reasoning and mathematical problems.

## **4. Tree of Thoughts (ToT):**
Multiple reasoning paths are generated and evaluated. This approach is useful for complex reasoning and planning problems.

## **5. Interview Approach:**
The model asks clarifying questions before answering. It helps in understanding ambiguous tasks and gathering more information before producing the final solution.

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="google/flan-t5-large",
    max_new_tokens=200
)

def generate(prompt):
    result = generator(prompt)
    return result[0]['generated_text']
```

```python
prompt_zero = """
Solve the following problem.

If a train travels 60 km in 1 hour, how far will it travel in 5 hours?
"""

response_zero = generate(prompt_zero)

print("Zero Shot Prompt")
print(response_zero)
```

```python
prompt_few = """
Example 1
Question: If a car travels 50 km in 1 hour, how far will it travel in 2 hours?
Answer: 100 km

Example 2
Question: If a bike travels 30 km in 1 hour, how far will it travel in 3 hours?
Answer: 90 km

Question: If a train travels 60 km in 1 hour, how far will it travel in 5 hours?
Answer:
"""

response_few = generate(prompt_few)

print("Few Shot Prompt")
print(response_few)
```

```python
prompt_cot = """
A shop sells pencils for 5 rupees each.
A student buys 8 pencils.

Let's think step by step and calculate the total amount.
"""

response_cot = generate(prompt_cot)

print("Chain of Thought")
print(response_cot)
```

```python
prompt_tot = """
A farmer has 10 apples.
He gives 3 apples to one friend and 2 apples to another friend.

Generate different reasoning paths and determine how many apples remain.
"""

responses = generator(prompt_tot, num_return_sequences=3)

print("Tree of Thoughts")

for i, r in enumerate(responses):
    print("Path", i+1)
    print(r['generated_text'])
    print()
```

```python
prompt_interview = """
You are an AI interviewer.

Before giving the final answer, ask clarifying questions.

Problem:
A company wants to use AI to improve customer service.
What should they do?
"""

response_interview = generate(prompt_interview)

print("Interview Approach")
print(response_interview)
```

```python
results = {
    "Zero Shot": response_zero,
    "Few Shot": response_few,
    "Chain of Thought": response_cot,
    "Interview Approach": response_interview
}

for method, output in results.items():
    print("====================================")
    print(method)
    print("====================================")
    print(output)
    print()
```
