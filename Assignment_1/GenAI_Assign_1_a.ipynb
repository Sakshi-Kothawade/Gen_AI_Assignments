```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
```

```python
X, y_true = make_blobs(
    n_samples=500,
    centers=3,
    cluster_std=1.2,
    random_state=42
)

plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title("Generated Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
gmm = GaussianMixture(
    n_components=3,
    covariance_type='full',
    random_state=42
)

gmm.fit(X)
```

```python
labels = gmm.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
plt.title("GMM Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
print("Means:\n", gmm.means_)
print("\nCovariances:\n", gmm.covariances_)
print("\nMixing Weights:\n", gmm.weights_)
```

```python
# Generating new data points
X_new, _ = gmm.sample(200)

plt.scatter(X[:, 0], X[:, 1], alpha=0.4, label="Original Data")
plt.scatter(X_new[:, 0], X_new[:, 1], color='red', label="Generated Data")
plt.legend()
plt.title("Data Generated by GMM")
plt.show()
```
